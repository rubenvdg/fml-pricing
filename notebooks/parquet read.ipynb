```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_parquet("../results.parquet")
```

```python
for c in range(4):
    print(f"segment: {c}.")
    df[f"x_{c}"].hist()
    plt.xlim((0, 1))
    plt.show()
    df.loc[lambda df: df["x.ub"] >= 0, "x.ub"].hist()
    plt.show()
    df[lambda df: df["x.ub"] >= 0].plot.scatter(f"x_{c}", "x.ub")
    plt.show()
```
